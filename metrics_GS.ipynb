```python
import string
import re
from pymorphy2 import MorphAnalyzer
import pprint
import pandas as pd
from pprint import pprint
from nltk.util import ngrams
from collections import Counter 
from nltk.corpus import stopwords 
import nltk.data
```

```python
#Постфикс _pic имеют переменные, в которых хранятся данные по корпусу "Истории о подарках и катании на лыжах"
#Постфикс _sib имеют переменные, в которых хранятся данные по корпусу "Рассказы сибиряков о жизни"
#Постфикс _nds имеют переменные, в которых хранятс данные по корпусу "Рассказы о сновидениях"
```

```python
#Читаем датасет:

df_pic = pd.read_csv('Pic.csv', encoding="utf-8", sep=",")
df_sib = pd.read_csv('Sib.csv', encoding="utf-8", sep=",")
df_nds = pd.read_csv('NDS.csv', encoding="utf-8", sep=",")
```

```python
#Убираем пустые строки:

df_pic = df_pic.dropna()
df_nds = df_nds.dropna() 
df_sib = df_sib.dropna() 
```

```python
#Берем наш текст из колонки Text:

pic = str(df_pic["Text"].values)
sib = str(df_sib["Text"].values.tolist()) #Затычка: пришлось применять tolist в данном датафрейме - str не дает всех значений здесь почему-то
nds = str(df_nds["Text"].values)
```

```python
#Функция, которая извлекает предложения из текста и подсчитывает их количество: 

def sents_extractor(text):
    text = re.sub(r'\.{3}', r'.',text) #Заменяем многоточие на точку, т к nltk.sent_tokenize плохо работает с многоточием
    text = nltk.sent_tokenize(text, language="russian")
    print("Количество предложений в корпусе: ", len(text))
    return text
```

```python
sents_pic = sents_extractor(pic)
```

```python
sents_sib = sents_extractor(sib)
```

```python
sents_nds = sents_extractor(nds)
```

```python
#Предобрабатываем текст: убираем пунктуацию, приводим все к нижнему регистру и расщепляем на слова:

def normalize(text):
    text = re.sub(r'[\[\]\n“”—!"\',\.:;\?`―\\]', '', text)
    text = text.lower().split()
    return text
```

```python
normalized_pic = normalize(pic)
#normalized_pic
```

```python
normalized_sib = normalize(sib)
#normalized_sib
```

```python
normalized_nds = normalize(nds)
#normalized_nds
```

```python
print ("Количество словоформ в корпусе 'Рассказы о подарках и лыжах':", len(normalized_pic))
print ("Количество словоформ в корпусе 'Рассказы сибиряков':", len(normalized_sib))
print ("Количество словоформ в корпусе 'Рассказы о сновидениях':", len(normalized_nds))
```

```python
#Функция, которая извлекает набор уникальных словоформ и подсчитывает их количество: 

def unique_wordforms (tokens):
    tokens = sorted(set(tokens))
    print("Количество уникальных словоформ: ", len(tokens))
    return tokens
```

```python
unique_wordforms_pic = unique_wordforms(normalized_pic)
unique_wordforms_pic
```

```python
unique_wordforms_sib = unique_wordforms(normalized_sib)
unique_wordforms_sib
```

```python
unique_wordforms_nds = unique_wordforms(normalized_nds)
unique_wordforms_nds
```

```python
from pymorphy2 import MorphAnalyzer

#сохраняем класс в переменную: 
morph = MorphAnalyzer() 
```

```python
#Функция, осуществляющая морфологический парсинг:

def morph_parse (tokens):
    words_analized = [morph.parse(word) for word in tokens]
    return words_analized
```

```python
morphparse_pic = morph_parse(normalized_pic)
#pprint(morphparse_pic)
```

```python
morphparse_sib = morph_parse(normalized_sib)
#pprint(morphparse_sib)
```

```python
morphparse_nds = morph_parse(normalized_nds)
#pprint(morphparse_nds)
```

```python
#Функция, извлекающая набор уникальных лемм из текста:

def unique_lemmas_extractor(tokens):
    lemmas_list=[]
    for t in tokens:
        lemmas_list.append(t[0].normal_form)
    unique_lemmas = set(lemmas_list)
    print("Количество уникальных лемм: ", len(unique_lemmas))
    return unique_lemmas
        
```

```python
unique_lemmas_pic = unique_lemmas_extractor(morphparse_pic)
pprint(unique_lemmas_pic )
```

```python
unique_lemmas_sib = unique_lemmas_extractor(morphparse_sib)
pprint(unique_lemmas_sib)
```

```python
unique_lemmas_nds = unique_lemmas_extractor(morphparse_nds)
pprint(unique_lemmas_nds)
```

```python
#Функция, извлекающая все части речи из текста:

def pos_extractor(tokens):
    pos_list = []
    for t in tokens:
        pos_list.append(t[0].tag.POS)
    return pos_list
```

```python
pos_list_pic = pos_extractor(morphparse_pic)
```

```python
pos_list_sib = pos_extractor(morphparse_sib)
```

```python
pos_list_nds = pos_extractor(morphparse_nds)
```

```python
#Функция, которая подсчитывает, сколько конкретных частей речи в тексте:

def pos_number_analysis(tokens):
    dct = {}
    for p in tokens:
        if p in dct:
            dct[p]+=1
        else:
            dct[p]=1
    return dct
```

```python
dct_pic = pos_number_analysis(pos_list_pic)
pprint(dct_pic)
```

```python
dct_sib = pos_number_analysis(pos_list_sib)
pprint(dct_sib)
```

```python
dct_nds = pos_number_analysis(pos_list_nds)
pprint(dct_nds)
```

```python
#Функция, вычисляющая все N-граммы из слов текста: 

def ngrams_by_nltk(tokens, num): 
    stop_words = stopwords.words('russian')+['ээ']+['эээ']+['мм'] #добавляем заполнители пауз в список стоп-слов
    tokens = [token for token in tokens if token not in stop_words]
    output = list(ngrams(tokens, num)) 
    return output
```

```python
ngrams_pic = ngrams_by_nltk(normalized_pic,2)

# Ищем 20 самых частотных биграмм в корпусе "Рассказы о подарках и катании на лыжах":
pic_top20 = Counter(ngrams_pic).most_common(20)
pprint(pic_top20)
```

```python
ngrams_sib = ngrams_by_nltk(normalized_sib,2)

# Ищем 20 самых частотных биграмм в корпусе "Рассказы сибиряков":
sib_top20 = Counter(ngrams_sib).most_common(20)
pprint(sib_top20)
```

```python
ngrams_nds = ngrams_by_nltk(normalized_nds,2)

# Ищем 20 самых частотных биграмм в корпусе "Рассказы сибиряков":
nds_top20 = Counter(ngrams_nds).most_common(20)
pprint(nds_top20)
```
